```python
import os
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import warnings
warnings.filterwarnings('ignore')
```

### Scrape programs in directory to compile dictionary

```python
# create the initial nested dictionary of with program number, operation number, and line of code

cwd = os.getcwd()
cwd = cwd.split('\\')[:5]
cwd = cwd[0] + '\\' + cwd[1] + '\\' + cwd[2] + '\\' + cwd[3] + '\\' + cwd[4] + '\\Programs\\CNHC'

root = (cwd)

# create empty dictionary as a storage container
program_dict = {}

# create variables for the path, subdirectories and files in 'root'
for path, subdirs, files in os.walk(root + '\\'):
    # iterate through the files
    for file in files:
        # create an emply dictionary to load G-code
        program_list = {}
        # create initial dictionary key for before N lines hit
        idx_name = 'Heading'
        # create a path to the file
        src = os.path.join(path, file)
        # create a variable to open the files
        o = open(src, 'r') # , encoding='Latin-1'
        # iterate over each line of the file
        for line in o.readlines():
            # if the line begins with N
            if line.startswith('N'):
                # capture the N number from the line
                N = line.split('(')[0]
                # rename the index name as the N number
                idx_name = N
            # if the N number is already a key in the dictionary, append the line
            if idx_name in program_list:
                program_list[idx_name] += line
            # else add the new key and line to the dictionary
            else:
                program_list[idx_name] = line
            # create nested dictionary using program (file name) as the new high level key
            program_dict[file] = program_list     
```

```python
# example of how to iterate through nested dict:

for program, dict2 in program_dict.items():
    program = program.split('-')[0]
    print(program)
    for key, block in dict2.items():
        print('    ' + key)
        block = block.split('\n')
        for line in block:       
            print('        ' + line)
```

### Capture specific data from dictionary

```python
# capture the data

# dictionary container
data = {}
k = float(0.01)
for program, dict2 in program_dict.items():
    program = program.split('-')[0]
    #print(program)
    temp_dict = {}
    for key, block in dict2.items():
        #print('    ' + key)
        # creates index on the block of code
        block = block.split('\n')
        # empty lists to store data
        feed_list = []
        tool_list = []
        spindle_list = []
        # empty dictionary to store data
        params = {}
        # for each line of code in the code block
        for line in block: 
            
            # TOOL
            
            # find the charcter T in a line
            if 'T' in line:
                # capture everything after the T
                tool = line.split('T')[1]
                #print(tool)
                if ',' in tool:
                    tool = tool.split(',')[0]
                if 'M13' in tool:
                    tool = tool.split('M13')[0]
                if 'M3' in tool:
                    tool = tool.split('M3')[0]
                # capture just the first characters
                #tool3 = tool[:3]
                #print(tool)
                tool = tool.replace(" ", "")
                if len(tool) == 3:
                    tool = '0' + tool
                #print(tool)
                #print(len(tool))
                if tool.isdigit():
                    if tool != '0':
                        print(tool)
                        tool_list.append(tool)
            
            # FEED
            
            # with F is in the line
            if 'F' in line:
                # capture everthing after F
                feed = line.split('F')[1]
                try:
                    feed = float(feed)
                    # update the feed list
                    feed_list.append(feed)
                except:
                    try:
                        feed = int(feed)
                        # update the feed list
                        feed_list.append(feed)
                    except:
                        pass
                #print(isinstance(feed, float) or isinstance(feed, int))
                
            # SPINDLE SPEEDS
            spindle_commands = ['G96S',
                                'G97S',
                                'G50S',
                                'G97G99S']
            for comm in spindle_commands:
                if comm in line:
                    try:
                        RPM = line.split(comm)[1]
                        RPM = RPM.split('M13')[0]
                        command = comm + RPM
                        spindle_list.append(command)
                    except:
                        pass
            # update the third dictionary, params, with found cutting parameters
            params.update({'Feed' : feed_list, 'Tool' : tool_list, 'Spindle Speed' : spindle_list})
        # update the second dictionary with the keys (N numbers) and parameter dict
        temp_dict.update({key : params})
    # update the first dictionary with program number and temporary dictionary of keys and parameters
    data.update({program : temp_dict})   
```

### Change dictionary to dataframe

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# create a Pandas dataframe from the nested dictionary 'data'
scraped_program_data = pd.DataFrame.from_dict({(i,j): data[i][j] 
                           for i in data.keys() 
                           for j in data[i].keys()},
                       orient='index').reset_index().rename(
    {'level_0': 'Program', 'level_1': 'Operation Number'}, axis=1)
# move brackets from list of lower level dictionary
cols = ['Feed',
        'Tool',
        'Spindle Speed']
for col in cols:
    scraped_program_data[col] =  scraped_program_data[col].astype(str)
    scraped_program_data[col] =  scraped_program_data[col].apply(lambda x: x.replace('[','').replace(']',''))
    scraped_program_data[col] =  scraped_program_data[col].apply(lambda x: x.replace("'",''))
# replace empty strings with NaN
scraped_program_data = scraped_program_data.replace(r'^\s*$', np.nan, regex=True)
# 
scraped_program_data['Operation Number'] = scraped_program_data['Operation Number'].apply(lambda x: x.replace('\n',''))
scraped_program_data.to_csv('scraped_program_data.csv', index=False)
# show the df
scraped_program_data
```
