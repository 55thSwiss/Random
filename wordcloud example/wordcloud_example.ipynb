```python
# generate the plot
d = getcwd()
mask = np.array(Image.open(path.join(d, "Untitled.png")))
wc = WordCloud(background_color="white", max_words=1000, mask=mask,
               max_font_size=60, random_state=3)
wc.generate(text)
# create coloring from image
image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[16,16])
plt.imshow(wc.recolor(color_func=image_colors),interpolation="bilinear")#interpolation="bilinear"
plt.axis("off")
_=plt.show()
```
