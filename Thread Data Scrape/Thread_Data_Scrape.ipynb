# G76 Program Scraping

```python
# import dependencies
import os

# create an empty dictionary
G76_dict = {}
# define root directory to search
root = r'\\REST-SAVEIT\\Users\\TaylorR\\OLD SAVEIT2\\CNC Master File\\ALL SWISS MACHINES\\'
# use 'os.walk' to check sub directories
for path, subdirs, files in os.walk(root + '\\'):
        # iterate thru checking the file names
        for name in files:
                # join the file name with it's raw path into variable 'directory'
                directory = os.path.join(path, name)
                # every program will have two G76 lines, this will start to address that
                new_name = (name + '_block_1')
                # all 'if not statements' below are checking file types just to save time in iteration
                if not directory.lower().endswith('.pdf'):
                        if not directory.lower().endswith('.doc'):
                                if not directory.lower().endswith('.part'):
                                        if not directory.lower().endswith('.ppt'):
                                                if not directory.lower().endswith('.dbp'):
                                                        if not directory.lower().endswith('.back'):
                                                                if not directory.lower().endswith('.cpd'):
                                                                        if not directory.lower().endswith('.trf'):
                                                                                if not directory.lower().endswith('.db'):
                                                                                        # open the directory's file in read only mode
                                                                                        o = open(directory, 'r', encoding='Latin-1')
                                                                                        # splitting the directory variable to catpure machine type
                                                                                        s = os.path.split(directory)
                                                                                        a = s[1]
                                                                                        b = s[0]
                                                                                        # same for drawing number
                                                                                        s = os.path.split(b)
                                                                                        c = s[1] 
                                                                                        b = s[0]
                                                                                        s = os.path.split(b)
                                                                                        # same for program number
                                                                                        e = s[1]
                                                                                        b = s[0]
                                                                                        # read through the lines of text one by one
                                                                                        for line in o.readlines():
                                                                                                # make the line lower case for processessing
                                                                                                line = str(line.lower())
                                                                                                line = line.strip().lower()
                                                                                                # look for string
                                                                                                if 'g76' in line:
                                                                                                        # if first G76 line already added to the dictionary
                                                                                                        # then change to 'block 2'
                                                                                                        if new_name in G76_dict:
                                                                                                                G76_dict.update({name + '_block_2' : [line, c, e]})
                                                                                                        else:
                                                                                                                # else add 'block 1'
                                                                                                                G76_dict.update({new_name : [line, c, e]})

# same as function above but with a different root
root = r'\\REST-SAVEIT\\Users\\TaylorR\\OLD SAVEIT2\\CNC Master File\ALL HARDINGE\\'
for path, subdirs, files in os.walk(root + '\\'):
        for name in files:
                directory = os.path.join(path, name)
                new_name = (name + '_block_1')
                if not directory.lower().endswith('.pdf'):
                        if not directory.lower().endswith('.doc'):
                                if not directory.lower().endswith('.part'):
                                        if not directory.lower().endswith('.ppt'):
                                                if not directory.lower().endswith('.dbp'):
                                                        if not directory.lower().endswith('.back'):
                                                                if not directory.lower().endswith('.cpd'):
                                                                        if not directory.lower().endswith('.trf'):
                                                                                if not directory.lower().endswith('.db'):
                                                                                        o = open(directory, 'r', encoding='Latin-1')
                                                                                        s = os.path.split(directory)
                                                                                        a = s[1]
                                                                                        b = s[0]
                                                                                        s = os.path.split(b)
                                                                                        c = s[1] 
                                                                                        b = s[0]
                                                                                        s = os.path.split(b)
                                                                                        e = s[1]
                                                                                        b = s[0]
                                                                                        for line in o.readlines():
                                                                                                line = str(line.lower())
                                                                                                line = line.strip().lower()
                                                                                                if 'g76' in line:
                                                                                                        if new_name in G76_dict:
                                                                                                                G76_dict.update({name + '_block_2' : [line, c, e]})
                                                                                                        else:
                                                                                                                G76_dict.update({new_name : [line, c, e]})
# write the dictionary to a .csv file
myfile = 'file.csv'
with open(myfile, 'w') as f:
        for key, value in G76_dict.items():
                f.write('%s, %s\n' % (key, value))
```

### Format 'Machine' Column
- There's a ton of columns that do not equal a real machine. This is from the windows explorer directory at 'root' above be structured differently from part to part. 
- For this exerise it will not be corrected. There will be fall out of data throughout the cleaning process and this is the first.

```python
#List unique values in the'Machine' column
df.Machine.unique()
```

```python
# create a list of real machine folder names from above
machineList = [" 'CHNC III']",
               " 'GT ROBOT']",
               " 'GT']",
               " 'KNC-32']",
               " 'NTY_3 Files']",
               " 'SA-16']",
               " 'SV-20']",
               " 'T-42']",
               " 'T-BO-326 TSUGAMI']",
               " 'T-S-206 TSUGAMI']",
               " 'TWIN TURN']",
               " 'WY-100']",
               " 'WY-150]'", ]
# remove any rows that don't contain a value from the list above in 'Machine'
df = df[df['Machine'].isin(machineList)]
# rename the 'Machine' string values to what they're supposed to be
df['Machine'] = df['Machine'].replace({" 'CHNC III']":'CHNC III', 
                                       " 'GT ROBOT']": 'GT ROBOT', 
                                       " 'GT']": 'GT',
                                       " 'KNC-32']": 'KNC-32',
                                       " 'NTY_3 Files']": 'NTY-3',
                                       " 'SA-16']": 'SA-16',
                                       " 'SV-20']": 'SV-20',
                                       " 'T-42']": 'T-42',
                                       " 'T-BO-326 TSUGAMI']": 'BO-326',
                                       " 'T-S-206 TSUGAMI']": ' S-206',
                                       " 'TWIN TURN']": 'TWIN TURN',
                                       " 'WY-100']": 'WY-100',
                                       " 'WY-150]'": 'WY-150'})
# check the output
df.Machine.unique()
```

```python
# looks like we lost about 400 rows of data, or 200 G76 blocks so far
df.shape
```

```python
# assign 'Block' column value based on pattern match from 'Program Number'
block1 = df['Program Number'].str.contains('block_1')
# replace the boolian return of the above line with '1' or '2' for block
df['Block'] = block1.replace((True,False), ('1','2'))
df.head(6)
```

```python
# clean the 'Drawing Number' columnn
char_list = ["'",
             ' REV',
             '.',]
# loop through the list above
for char in char_list:
    # replace the character with nothing, esseentially deleting it
    df['Drawing Number'] = df['Drawing Number'].str.replace(char, '')
# replace other strings
df['Drawing Number'] = df['Drawing Number'].str.replace('REV-', ' ')
df['Program Number'] = df['Program Number'].str.replace('_block_1', '')
df['Program Number'] = df['Program Number'].str.replace('_block_2', '')
# clean 'G76 Code' column
char_list = ['[',
             "'",
             '/',]
for char in char_list:
    df['G76 Code'] = df['G76 Code'].str.replace(char, '')
# uppcase characters in 'G76 Code' column
df['G76 Code'] = df['G76 Code'].str.upper()
```

```python
# iterate the df
for index, row in df.iterrows():
    # find a '-' in 'Drawing Number' as 5th character, for strings like "xxxx-xxxx DTCD0100070 A"
    if row[2].find('-') == 5:
        # remove the unwanted leading characters
        row[2] = row[2][10:]
    # remove the whitespace before and after string in 'Program Name':
    row[2] = row[2].rstrip(' ')
    row[2] = row[2].lstrip(' ')
df.head()
#df.shape
```

```python
# grouping every two rows based on Program Number, Drawing Number, and Machine to get the two G76 lines on the same row of data
df = df.groupby('Program Number').agg({'Drawing Number':'first', 
                             'G76 Code': ', '.join, 
                             'Machine':'first' }).reset_index()
df.head()
```

```python
df.shape
```

### Merge data frames

```python
# merge the two dataframes, scraped data & MS Access table data
# they'e being merged on the 'Drawing Number' column, this is another step that will
# have a lot of fall out due to the directory inconsistency from earlier
df_new = pd.merge(df, dfMaterial, how = 'left', on = 'Drawing Number')
# show a sample
df_new.sample(6)
```

```python
df_new.shape
```

### Final cleaning and formatting

```python
# drop any rows that contain a 'NaN' value (not a number)
df_new.dropna(inplace=True)
df_new.shape
```

```python
df_new.head(251)
```

```python
# iterate over the dataframe
for index, row in df_new.iterrows():
    # strip white space on left of row value
    row[2] = row[2].lstrip(' ')
    # if the G76 column doesn't start with G
    if (str(row[2])[0]) != 'G':
        # drop if it
        df_new.drop(index, inplace = True)
# because of initial rows being dropped, certain observations could not be found. Required it's own iterrow()
for index, row in df_new.iterrows():        
    # check for ',' in G76 Code column
    if ',' not in str(row[2]):
        # drop if it doesn't exist
        df_new.drop(index, inplace = True)
df_new.shape
```

```python
# split 'G76 Code' column by ','
split = df_new['G76 Code'].str.split(',', n = 1, expand = True) 
# set first split to new column 
df_new['G76 Block 1']= split[0] 
# set second split to second new column
df_new['G76 Block 2']= split[1] 
# drop the original column
df_new.drop(columns =['G76 Code'], inplace = True) 
# reorganize columns
df_new = df_new[[ 'Drawing Number', 'Material', 'Program Number', 'G76 Block 1', 'G76 Block 2', 'Machine']]
df_new.head(233)
```

### Output data to CSV

```python
df_new.to_csv('G76_Data.csv')
```
